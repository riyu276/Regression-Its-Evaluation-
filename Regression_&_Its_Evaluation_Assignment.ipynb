```python
Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Program=>

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Reshape to 2D array
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Print slope and intercept
print("Intercept (β0):", model.intercept_)
print("Slope (β1):", model.coef_[0])

# Plot data points and regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.

Program =>

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (y)
X = df[["Area", "Rooms"]]
y = df["Price"]

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print("Intercept (β0):", model.intercept_)
print("Coefficient for Area (β1):", model.coef_[0])
print("Coefficient for Rooms (β2):", model.coef_[1])

# Regression equation
print("\nEquation: Price = {:.2f} + {:.2f}*Area + {:.2f}*Rooms".format(
    model.intercept_, model.coef_[0], model.coef_[1]
))

# Check multicollinearity using VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

```python
Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
3
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.

Program =>

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform input for polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit Polynomial Regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predictions
Y_pred = model.predict(X_poly)

# Print coefficients
print("Intercept (β0):", model.intercept_)
print("Coefficients (β1, β2):", model.coef_[1:])

# Plot data and polynomial regression curve
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Polynomial Regression (Degree 2)')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.legend()
plt.show()
```

```python
Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.

Program =>

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit Linear Regression
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Print results
print("Predicted Values:", Y_pred)
print("Residuals:", residuals)

# Plot residuals
plt.scatter(Y_pred, residuals, color="blue", label="Residuals")
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.legend()
plt.show()
```

Question 1: What is Simple Linear Regression?

Answer :- Simple Linear Regression is a statistical method used to model the relationship between two variables:

* Independent variable (X): The input or predictor.

* Dependent variable (Y): The output or response we want to predict.

It assumes that the relationship between X and Y is linear, meaning it can be represented with a straight line:

Y=β0​+β1​X+ε

Where:

* Y = Dependent variable (target)

* X = Independent variable (predictor)

* 𝛽0 = Intercept (value of Y when X=0)

* β1 = Slope (change in Y for a one-unit change in X)

* ε = Error term (difference between actual and predicted values)

Goal:
To find the best-fitting straight line through the data points, usually by minimizing the sum of squared errors (SSE) between predicted and actual values.

Example:
If we want to predict a student’s exam score (Y) based on study hours (X), simple linear regression would try to find the line that best explains how scores increase with hours studied.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer: Here are the key assumptions of Simple Linear Regression:

1. Linearity

* The relationship between the independent variable (X) and the dependent variable (Y) is linear.

* Mathematically:
Y=β0+β1X+ε.

2. Independence of Errors

* The residuals (errors) are independent of each other.

* No autocorrelation (common in time series data).

3. Homoscedasticity (Constant Variance of Errors)

* The variance of residuals is constant across all values of X.

* In other words, the spread of errors should be roughly the same throughout the regression line.

4. Normality of Errors

* The residuals (errors) are normally distributed.

* This is especially important for hypothesis testing and confidence intervals.

5. No or Minimal Multicollinearity (only relevant in multiple regression)

* In simple linear regression, there’s only one predictor, so this assumption doesn’t apply.

* But in multiple regression, predictors shouldn’t be highly correlated.

Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

Answer: Heteroscedasticity means that the variance of the error terms (residuals) is not constant across all values of the independent variable(s).

Why it is important to address:

1. Biased Standard Errors

* Heteroscedasticity doesn’t bias the regression coefficients (𝛽0,𝛽1), but it does bias the standard errors.

* This makes statistical tests (like t-tests, F-tests) unreliable.

2.Incorrect Hypothesis Testing

* Because standard errors are biased, p-values can be misleading.

* You may incorrectly conclude that a variable is significant (Type I error) or insignificant (Type II error).

3. Reduced Model Efficiency

* Ordinary Least Squares (OLS) is no longer the Best Linear Unbiased Estimator (BLUE) when heteroscedasticity exists.

* Predictions may be less reliable.

Example:

If you are modeling the relationship between income (X) and amount spent on luxury goods (Y), the variability of spending is much higher for high-income groups compared to low-income groups. This creates heteroscedasticity.

Question 4: What is Multiple Linear Regression?

Answer: Multiple Linear Regression (MLR) is a statistical technique used to model the relationship between one dependent variable (Y) and two or more independent variables (X₁, X₂, …, Xₙ).

It extends simple linear regression (which has only one predictor) by considering multiple predictors simultaneously.

Mathematical Equation:

Y=β0​+β1​X1​+β2​X2​+⋯+βn​Xn​+ε

Where:

* Y = Dependent variable (target)

* X1,X2,…,Xn = Independent variables (predictors)

* 𝛽0 = Intercept (value of Y when all X’s are 0)

* 𝛽1,𝛽2,…,𝛽𝑛= Coefficients (showing how much Y changes with a one-unit change in each X, holding other variables constant)

* ε = Error term (unexplained variation in Y)

Key Points:

1. Captures complex relationships – Unlike simple regression, it can explain Y using multiple factors.

2. Assumptions – Same as linear regression (linearity, independence, homoscedasticity, normality of residuals, no multicollinearity).

3. Uses – Prediction, forecasting, and understanding the relative influence of different variables.

Example:

If you want to predict a house price (Y) based on:

* Size of the house (𝑋1)

* Number of bedrooms (𝑋2)

* Location score (𝑋3)

Then MLR would build a model like:

House Price = β0​+β1​(Size)+β2​(Bedrooms)+β3​(Location)+ε

Question 5: What is polynomial regression, and how does it differ from linear regression?

Answer: Polynomial Regression

Polynomial Regression is a type of regression analysis where the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth-degree polynomial.

It is an extension of linear regression that allows us to capture non-linear relationships.

Equation:

For a polynomial of degree 2 (quadratic regression):

Y=β0​+β1​X+β2​X2+ε

For a general nth-degree polynomial:

Y=β0​+β1​X+β2​X2+β3​X3+⋯+βn​Xn+ε

Here’s the difference between Polynomial Regression and Linear Regression

1. Nature of Relationship

* Linear Regression shows a straight-line relationship between independent and dependent variables.

* Polynomial Regression shows a curved or non-linear relationship by including higher powers of the independent variable.

2. Equation

* Linear Regression: Y=β0+β1X+ε

* Polynomial Regression: Y=β0​+β1​X+β2​X2+⋯+βn​Xn+ε

3.Flexibility

* Linear Regression is less flexible and works only for linear trends.

* Polynomial Regression is more flexible and fits complex curved patterns.

4. Interpretability

* Linear Regression is simple to interpret with slope and intercept.

* Polynomial Regression becomes harder to interpret as the polynomial degree increases.

5. Risk of Overfitting

* Linear Regression has a low risk of overfitting.

* Polynomial Regression has a high risk of overfitting, especially with high degrees.

6. Use Cases

* Linear Regression is used when data follows a straight-line trend.

* Polynomial Regression is used when data shows non-linear or curved trends.

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location.However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust model.

Answer: To build a robust regression model for predicting house prices, it is important to address both heteroscedasticity and multicollinearity issues. The following steps can be taken:

1.Addressing Heteroscedasticity

* Residual Plot Analysis: First, confirm heteroscedasticity by plotting residuals against fitted values.

* Data Transformation: Apply transformations such as logarithmic, square root, or Box-Cox on the dependent variable (house prices) to stabilize variance.

* Weighted Least Squares (WLS): Assign weights to observations so that those with higher variance get less weight.

* Robust Standard Errors: Use heteroscedasticity-consistent standard errors (e.g., White’s standard errors) to make hypothesis testing reliable.

2. Addressing Multicollinearity

* Variance Inflation Factor (VIF): Calculate VIF to identify highly correlated predictors. A VIF greater than 10 usually indicates multicollinearity.

* Remove or Combine Features: Drop redundant variables (e.g., if both "total rooms" and "bedrooms" are highly correlated, keep only one).

* Feature Engineering: Create new features that combine correlated predictors meaningfully, such as "price per square foot."

* Regularization Techniques: Apply methods like Ridge Regression or Lasso Regression, which penalize large coefficients and reduce the effect of multicollinearity.

3. Ensuring a Robust Model

* Cross-Validation: Use k-fold cross-validation to ensure the model generalizes well on unseen data.

* Model Comparison: Compare performance with alternative models (like decision trees or random forests) that are less sensitive to multicollinearity and heteroscedasticity.

* Interpretability: After adjustments, interpret coefficients carefully, ensuring predictors kept in the model are meaningful and stable.
